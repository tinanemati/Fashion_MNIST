## **Summary** 
This program designs a neural network that can classify images of clothes with at least 91% accuracy to assist in the return process at a shipping facility. Our input dataset consists a training set of 60,000 examples and a test set of 10,000 examples. Each example is a 28x28 grayscale image, associated with a label from 10 classes.  We have calculated that the network should be at least 91% accurate to ensure this sorting process is beneficial.

##**Architecture of our network**
We built our neaural network using PyTorch. Our goal here was to classify 10 classes apparel images in the Fashion-MNIST dataset with at least 91%  accuracy. We took advantage of the in-built Fashion-MNIST dataset from PyTorch's `torchvision` package in order to make our life easier when we iterate through the image samples while training and testing the model. 

In the design process of our neural network, we took advantage of the standard deep learning architecture given in the slides and repesented bellow: 

In our implementation of standard design, we have the followings to help us processing the output: 
- 2 convalutional layers that contains a set of kernals. 
  - allowed us to have a fairly simple design instead of a complex one 
  - helped us simplify image and computation without loosing too much  
  information as we continue. 
- maxpooling to help us even further in simplifying our data in each layer. 
  - calculated the maximum value in each of feature map.  
- 5 linear layers with the dropout rate of 0.15. 
  - forced our network to not rely on any single neuron by randomly zero out some input of the outgoing activations.
  - prevented overfitting in our network.
- Used Relu Non Linearity for the model and logsoftmax at the last layer.


##**Tuning of our network and Performance**
As mentioned before, our neaural network consists of 2 convalutional layers and 5 linear layers. At first, we tried to increase the linear layers as well as the epoch times for improving the accuracy . However, we found out the accuracy was decreased by doing so because the over training occurs easily when epoch times is increased. Then, to figure out this issue, we discussed with one of our TAs if there is another solution that could increase the accuracy without setting up too many linear layers. With the help of the TA, we finally decided to add some dropout layers with the default rate of 0.5 instead of linear layers and the accuracy was up to 87%. By gaining some insights from the dropout layers, we lowered the probability of dropout to be 0.15 and tried to set different epoch times to make sure our accuracy could be stable over 91%. 

Finally, with these changes implemented and the epoch of time of 100, we were able to fully reach the goal for our network which can be shown in the test results bellow:

# PyTorch

## 0 Imports

```python
import torch
import torchvision
import torchvision.transforms as transforms
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import matplotlib.pyplot as plt

device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
```

## 1. Set Hyperparameters

```python
# increased n_epoch time value to 100 to reach 91% accuracy. 

n_epochs = 100
batch_size_train = 64
batch_size_test = 1000
learning_rate = 0.01
momentum = 0.5
log_interval = 100
```

## 2. Load Data

```python
train_set = torchvision.datasets.FashionMNIST(
  root = './data/FashionMNIST',
  train = True,
  download = True,
  transform = transforms.Compose([
      transforms.ToTensor()                                 
  ]))

train_loader = torch.utils.data.DataLoader(train_set,
                                     batch_size = batch_size_train,
                                     shuffle=True)

train_set = torchvision.datasets.FashionMNIST(
  root = './data/FashionMNIST',
  train = False,
  download = True,
  transform = transforms.Compose([
      transforms.ToTensor()                                 
  ]))

test_loader = torch.utils.data.DataLoader(train_set,
                                     batch_size = batch_size_train,
                                     shuffle=True)
```

```python
examples = enumerate(test_loader)
batch_idx, (example_data, example_targets) = next(examples)

fig = plt.figure()
for i in range(6):
  plt.subplot(2,3,i+1)
  plt.tight_layout()
  plt.imshow(example_data[i][0], cmap='gray', interpolation='none')
  plt.title("Ground Truth: {}".format(example_targets[i]))
  plt.xticks([])
  plt.yticks([])
```

```python
# changed INPUT_CHANNELS to example_data.shape[1]. 
INPUT_CHANNELS = example_data.shape[1]
OUTPUT_CLASSES = 10
example_data.shape
```

## 3. Define Model

```python
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        # two convolution layers 
        self.conv1 = nn.Conv2d(INPUT_CHANNELS, 10, 5, 1)
        self.conv2 = nn.Conv2d(10, 20, 5, 1)
        # max pooling to simplify data 
        self.maxPooling = nn.MaxPool2d(2)
        # five linear layers with 0.15 dropout rate
        self.fc1 = nn.Linear(320, 256)
        self.fc2 = nn.Linear(256, 192)
        self.fc3 = nn.Linear(192, 128)
        self.fc4 = nn.Linear(128, 64)
        self.fc5 = nn.Linear(64, OUTPUT_CLASSES)
        self.dropout1 = nn.Dropout(0.15)
        self.dropout2 = nn.Dropout(0.15)
        self.dropout3 = nn.Dropout(0.15)
        self.dropout4 = nn.Dropout(0.15)
        self.dropout5 = nn.Dropout(0.15)
    
  


    def forward(self, x):
        x = self.conv1(x)
        x = self.maxPooling(x)
        x = self.conv2(x)
        x = self.maxPooling(x)
        linearize = nn.Flatten()
        x = linearize(x)
        x = self.dropout1(F.relu(self.fc1(x)))
        x = self.dropout2(F.relu(self.fc2(x)))
        x = self.dropout3(F.relu(self.fc3(x)))
        x = self.dropout4(F.relu(self.fc4(x)))
        x = self.dropout5(F.relu(self.fc5(x)))
       
        output = F.log_softmax(x, dim=1)
        return output
```

```python
net = Net().to(device)
net
```

## 4. Train Model

```python
train_losses = []
train_counter = []
test_losses = []
test_counter = [i*len(train_loader.dataset) for i in range(n_epochs + 1)]

optimizer = optim.SGD(net.parameters(), lr=learning_rate, momentum=momentum)
criterion = nn.MSELoss()
```

```python
def train(model, train_loader, optimizer, epoch):
    model.train()
    for batch_idx, (data, target) in enumerate(train_loader):
        data = data.to(device)
        target = target.to(device)
        optimizer.zero_grad()
        output = model(data)
        loss = F.nll_loss(output, target)
        loss.backward()
        optimizer.step()
        if batch_idx % log_interval == 0:
            print('Train Epoch: {} [{}/{} ({:.0f}%)]\tLoss: {:.6f}'.format(
                epoch, batch_idx * len(data), len(train_loader.dataset),
                100. * batch_idx / len(train_loader), loss.item()))


def test(model, test_loader):
    model.eval()
    test_loss = 0
    correct = 0
    with torch.no_grad():
        for data, target in test_loader:
            data = data.to(device)
            target = target.to(device)
            output = model(data)
            test_loss += F.nll_loss(output, target, reduction='sum').item()  # sum up batch loss
            pred = output.argmax(dim=1, keepdim=True)  # get the index of the max log-probability
            correct += pred.eq(target.view_as(pred)).sum().item()

    test_loss /= len(test_loader.dataset)

    print('\nTest set: Average loss: {:.4f}, Accuracy: {}/{} ({:.0f}%)\n'.format(
        test_loss, correct, len(test_loader.dataset),
        100. * correct / len(test_loader.dataset)))
```

```python
for epoch in range(1, n_epochs + 1):
  train(net, train_loader, optimizer, epoch)
  test(net, test_loader)
```
